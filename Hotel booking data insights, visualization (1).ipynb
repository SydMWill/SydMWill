```python
#Hotel booking data insights, visualization
import pandas as pd

df = pd.read_csv('/Users/sydney/Downloads/hotel_bookings.csv')
df
abbreviated_df= df.head(20)
print(abbreviated_df)
#Data frame from https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand?resource=download
#This dataset focuses on the booking data for resort and city hotels, however, the first 20 rows are only resorts.
#head() gives the number of inputted rows for a given data frame.
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html

#Analyzing canceled bookigs
canceled_bookings = df.head(20)['is_canceled'].sum()
non_canceled_bookings = 20 - canceled_bookings

# Print the results
print("Number of canceled bookings:", canceled_bookings)
print("Number of non-canceled bookings:", non_canceled_bookings)

#Calculating the number of cancellations vs. non-cancellations gives us insight into
#the rate of cancellations. From this portion of the dataset, I would hypothesize that in July
#cancellation rates for resorts are low. July is a popular travel month that people look forward to all year.

#Analyze stays during weekends vs week nights
avg_weeknight_stays = df.head(20)['stays_in_week_nights'].mean()
avg_weekend_stays = df.head(20)['stays_in_weekend_nights'].mean()

# Print the results
print("Average number of stays during weeknights:", avg_weeknight_stays)
print("Average number of stays during weekend nights:", avg_weekend_stays)


#Another hypothesis, is that given the amount of reservations
#made during week nights,I would suggest that people are reserving for business trips or taking time off for leisure

```

```python
import matplotlib.pyplot as plt

# Plotting canceled vs non-canceled bookings
plt.figure(figsize=(10, 5))
plt.bar(['Canceled', 'Non-Canceled'], [canceled_bookings, non_canceled_bookings], color=['red', 'green'])
plt.title('Canceled vs Non-Canceled Bookings')
plt.ylabel('Number of Bookings')
plt.show()

# Plotting average stays during weeknights vs weekend nights
plt.figure(figsize=(10, 5))
plt.bar(['Weeknights', 'Weekend nights'], [avg_weeknight_stays, avg_weekend_stays], color=['blue', 'orange'])
plt.title('Average Stays during Weeknights vs Weekend Nights')
plt.ylabel('Average Number of Stays')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#seaborn visualizes random distribution
#https://www.w3schools.com/python/numpy/numpy_random_seaborn.asp
# Filter lead time for the first 20 rows
lead_time_data = df.head(20)['lead_time']

# Plotting lead time distribution
plt.figure(figsize=(10, 6))
sns.histplot(lead_time_data, kde=True, color='blue', bins=20)
plt.title('Lead Time Distribution')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Frequency')
plt.show()
```
